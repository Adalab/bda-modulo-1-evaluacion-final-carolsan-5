```python
# Función 1 -

def agregar_producto(nombre, precio, cantidad):
    
    for producto in inventario_productos:
        if producto["nombre"] == nombre:
            producto["cantidad"] += cantidad                      # si el producto se encuentra en el inventario, entonces actualiza la cantidad
            return "El producto ya existe en el inventario, se ha actualizado la cantidad"
        
    inventario_productos.append({"nombre": nombre,"precio": precio,"cantidad": cantidad})
    return "No teníamos el producto, así que lo hemos añadido"
```

```python
# Función 2 -

def ver_inventario():
    inventario = []

    for producto in inventario_productos:
         inventario.append(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
    print("\n".join(inventario))
```

```python
# Función 3 -

def buscar_producto(nombre):
    productos = []
    
    for producto in inventario_productos:
        if producto['nombre']== nombre:
            productos.append(producto)
    
    if len(productos) == 0:
        print(f"Prueba otra vez, ninguna coincidencia con {nombre}")

    else:
        return productos
```

```python
# Función 4 -

def actualizar_stock(nombre, cantidad):
    
    for producto in inventario_productos:
        if producto["nombre"] == nombre:
            producto["cantidad"] += cantidad     
            ver_inventario()               
            return "El producto ya existe en el inventario, se ha actualizado la cantidad"
        
    return "El producto no se encuentra en el inventario."
```

```python
# Función 5 -

def eliminar_producto(nombre):
    
    for producto in inventario_productos:
        if producto["nombre"] == nombre:
            inventario_productos.remove(producto)
            return f"Se ha eliminado el producto correctamente"
        
    print("Producto no encontrado, lo sentimos") 
```
